```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/DataSet/sample_sales_data.csv'
```

```python
df = pd.read_csv(path)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract year and month from the timestamp
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month

# Group by year, month, and category, then calculate total amount and total quantity
result = df.groupby(['year', 'month', 'category']).agg({
    'total': 'sum',
    'quantity': 'sum'
}).reset_index()
```

```python
result
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='category', y='total', hue='year', data=result)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plotting total quantity for each category
plt.figure(figsize=(12, 8))
sns.barplot(x='category', y='quantity', hue='year', data=result)
plt.title('Total Quantity by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
